## Visualize your path to financial independence, Race to Fire (Financial Independence, Retire Early)

```python
# Import the libraries
import matplotlib.pyplot as plt
```

```python
# Define a function to calculate the years until FIRE and track the savings growth
def calculate_fire_years(current_age, annual_expenses, savings_rate, annual_income, current_savings, expected_return = 0.07, 
                         withdrawl_rate = 0.04):
    fire_target = annual_expenses / withdrawl_rate  # FIRE target or FIRE number
    annual_savings = annual_income * savings_rate # Calculates how much is saved each year
    years = 0  # Years counter
    savings_history = [current_savings] # Start tracking savings over time

    # Loop until the savings reach or exceed the FIRE target
    while current_savings < fire_target:
        # Applying compound growth to current savings and adding the annual savings to the portfolio
        current_savings = current_savings * (1 + expected_return) + annual_savings  
        savings_history.append(current_savings)  # Recording the updated savings (for the plot)
        years += 1 # Incrementing the year counter

        # Safety check: prevent an infinite loop if FIRE is unreachable
        if years > 100:
            return None, None, savings_history

    fire_age = current_age + years  # Calculate the age when FIRE is reached

    return years, fire_age, savings_history
```

```python
# Define a function to plot the savings growth over time vs FIRE target
def plot_savings_growth(savings_history, fire_target):
    plt.figure(figsize=(10,6))
    plt.plot(savings_history, label='Savings Over Time', color='blue')
    plt.axhline(y=fire_target, color='red', linestyle='--', label='FIRE target')
    plt.xlabel('Years')
    plt.ylabel('Total Savings ($)')
    plt.title('FIRE Progression: Savings vs Target')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Get the users input and run the program
current_age = 20 # user's current age
annual_expenses = 17500  # estimated yearly spending
savings_rate = 0.5  # portion of income saved annually
annual_income = 35000 # total annual income
current_savings = 0 # current savings / investments
expected_return = 0.08
withdrawl_rate = 0.08

# Run FIRE calculation with the User's input
years, fire_age, savings_history = calculate_fire_years(current_age, annual_expenses, savings_rate, 
                                                        annual_income, current_savings, expected_return, withdrawl_rate)

# Display the results if FIRE is achievable 
if years != None:
    print("You can reach FIRE in ", years, "year(s) or less, at age", fire_age, ".")
    plot_savings_growth(savings_history, annual_expenses / 0.08)
else:
    print("FIRE goal may be unrealistic with current inputs.")
```
